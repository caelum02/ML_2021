```python
import numpy as np
import matplotlib.pyplot as plt
```

## Set random seed, hyperparameters, and some values

```python
# set random seed
np.random.seed(1337)

# regularization hyperparameter lambda
reg = 0.1
# number of data
n = 1000
# dimension of X
d = 100
# step size
lr = 0.01
# number of iteration
it = 100
```

## Generate dataset and initial weight

```python
X = np.vstack([np.random.normal(0.1, 1, (n//2, d)), np.random.normal(-0.1, 1, (n//2, d))])
y = np.hstack([np.ones(n//2), -1.*np.ones(n//2)])
w0 = np.random.normal(0, 1, d)
```

## Define essential functions

```python
def loss(w):
    return np.mean(np.maximum(1-y*(X@w), 0))+0.5*reg*w.T@w

def subgradient(w):
    return (((1-y*(X@w) > 0)*(-y)).reshape([-1, 1])*X).mean(axis=0) + reg*w
    
def accuracy(w):
    return np.mean(np.sign(X@w).squeeze()==y)
```

## Run Gradient Descent

```python
# set weight to inital weight
w = w0.copy()

losses = []
acc = []

# run gradient descent
for i in range(it):
    w -= lr*subgradient(w)
    
    losses.append(loss(w))
    acc.append(accuracy(w))

plt.plot(np.arange(it), losses)
plt.xlabel("# of iteration")
plt.ylabel("loss")
plt.show()
plt.plot(np.arange(it), acc)
plt.ylabel("accuracy")
plt.xlabel("# of iteration")
plt.show()
```
