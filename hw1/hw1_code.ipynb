```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 1000
p = 12
```

Run Apprioximate Gaussian Sample algorithm n times, and sample n gaussian samples.

```python
s = [np.random.random(p).sum()-6 for i in range(n)]
z = np.random.randn(n)
```

# Q2-2

```python
plt.hist(s, bins=20, alpha=0.5, density=True, label="approximate sample")
plt.hist(z, bins=20, alpha=0.5, density=True, label="gaussian sample")
plt.xlim(-5, 5)
plt.legend()
plt.plot()
```

# Q2-3

```python
from scipy.stats import gaussian_kde
```

```python
gaussian_pdf = lambda x: (2*np.pi)**-0.5*np.exp(-0.5*x**2)
x = np.arange(-5, 5, 0.001)
```

```python
plt.hist(s, bins=20, alpha=0.5, density=True, label="approx. hist")
plt.hist(z, bins=20, alpha=0.5, density=True, label="gaussian. hist")
plt.plot(x, gaussian_kde(s, bw_method='silverman')(x), label="approx. kde")
plt.plot(x, gaussian_kde(z, bw_method='silverman')(x), label="gaussian. kde")
plt.plot(x, gaussian_pdf(x), label="gaussian pdf")
plt.xlim(-5, 5)
plt.legend()
plt.plot()
```
