```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def simplex_projection(s):
    """Projection onto the unit simplex."""
    if np.sum(s) <=1 and np.alltrue(s >= 0):
        return s
    # Code taken from https://gist.github.com/daien/1272551
    # get the array of cumulative sums of a sorted (decreasing) copy of v
    u = np.sort(s)[::-1]
    cssv = np.cumsum(u)
    # get the number of > 0 components of the optimal solution
    rho = np.nonzero(u * np.arange(1, len(u)+1) > (cssv - 1))[0][-1]
    # compute the Lagrange multiplier associated to the simplex constraint
    theta = (cssv[rho] - 1) / (rho + 1.0)
    # compute the projection by thresholding v using theta
    return np.maximum(s-theta, 0)

def nuclear_projection(A):
    """Projection onto nuclear norm ball."""
    U, s, V = np.linalg.svd(A, full_matrices=False)
    s = simplex_projection(s)
    return U.dot(np.diag(s).dot(V))
```

```python
np.random.seed(1337)

n, k = 1000, 10
# random rank-10 matrix normalized to have nuclear norm 1
U = np.random.normal(0, 1, (n, k))
U = np.linalg.qr(U)[0]
S = np.diag(np.random.uniform(0, 1, k))
S /= np.sum(S)
A = U.dot(S.dot(U.T))
# pick which entries we observe uniformly at random
O = np.random.randint(0,2, (n, n))
# multiply A by O coordinate-wise
Y = np.multiply(A, O)

X0 = np.random.normal(0,1, (n,n))
X0 = nuclear_projection(X0.dot(X0.T))
```

```python
def objective(X):
    return 0.5 * np.linalg.norm(Y-O*X, ord='fro')**2

def grad(X):
    return O*X-Y

def step_size(t):
    return 2 / (t+2)

def power_method(M, num_iters=10):
    m, n = M.shape
    S = M@M.T # mxm
    u = np.random.random(m)
    u /= np.linalg.norm(u)
    for i in range(num_iters):
        u = S @ u
        u /= np.linalg.norm(u)
    v = M.T @ u
    v /= np.linalg.norm(v)
    return u, v

def linear_optimize(X):
    u, v = power_method(-grad(X))
    return u[:,None] @ v[None,:]
```

```python
MAX_STEP = 40
X = X0.copy()
loss = [objective(X)]

for t in range(1, MAX_STEP+1):
    X -= step_size(t) * grad(X)
    X = nuclear_projection(X)
    loss.append(objective(X))
```

```python
MAX_STEP = 40
X = X0.copy()
loss = [objective(X)]
np.random.seed(41)
for t in range(1, MAX_STEP+1):
    X += step_size(t) * (linear_optimize(X)-X)
    loss.append(objective(X))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(range(0, MAX_STEP+1), loss)
plt.yscale('log')
```

```python
S_true = np.linalg.svd(A, compute_uv=False,)
S_recovered = np.linalg.svd(X, compute_uv=False)
plt.plot(range(1, 21), S_true[:20], label="true")
plt.plot(range(1, 21), S_recovered[:20], label="recovered")
plt.legend()
plt.plot()
```
